```python
#Reference:http://hamelg.blogspot.tw/2015/12/python-for-data-analysis-part-30-random.html

import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing
```

```python
titanic_train = pd.read_csv("D:/Li-Sin/Documents/Jupyter/BigDataAnalytics/Microsoft-Malware-Challenge-master/Dataset/train/LargeTrain.csv")    # Read the data
np.random.seed(12)
label_encoder = preprocessing.LabelEncoder()
features = list(titanic_train.columns.values)
```

```python
rf_model = RandomForestClassifier(n_estimators=1000, 
                                  max_features=2,    
                                  oob_score=True)    

rf_model.fit(X=titanic_train[features],
             y=titanic_train["Class"])

print("OOB accuracy: ")
print(rf_model.oob_score_)
```

```python
score=[]
for feature, imp in zip(features, rf_model.feature_importances_):
    score.append(imp)
    print(feature, imp)
```

```python
df = pd.DataFrame(columns=['features','score'])
df['features']=features
df['score']=score
df = df.drop([1804]) #把Lable去掉
```
